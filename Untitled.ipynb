```python

class Context:
    def __init__(self, G, M, I):
        self.G = G
        self.I = I
        self.M = M
        inner = dict()
        for idxg in range(len(G)):
            row = dict()
            for idxm in range(len(M)):
                row[M[idxm]] = I[idxg][idxm]
            inner[G[idxg]] = row
        self.inner = inner
        
    def getRowByG(self, g):
        return self.inner[g]
    def getColByM(self,m):
        res = dict()
        for (key, val) in enumerate(self.inner):
            res[val] = self.inner[val][m]
        return res

class Node:
    def __init__(self, G, M):
        self.G = G
        self.M = M
    def __hash__(self):
        return hash(frozenset(self.G)) + hash(frozenset(self.M))
    def __eq__(self, other):
        if not isinstance(other, type(self)): return NotImplemented
        return self.G == other.G and self.M == other.M
    def __repr__(self):
        return"(" + ", ".join(self.G) + ") (" + ", ".join(self.M) + ")"

class Lattice:
    def __init__(self, C, E):
        self.C = C
        self.E = E
```

```python
from functools import reduce
    
def commonAtrs(ctx, G1, M1):
    res = set()
    for m in M1:
        col = ctx.getColByM(m)
        vals = []
        for key in col:
            if (key in G1):
                vals.append(col[key])
        if (0 not in vals):
            res.add(m)
    return res
def commonObjForAtt(ctx, G1, M1):
    res = []
    for m in M1:
        col = ctx.getColByM(m)
        vals = set()
        for key in col:
            if (key in G1 and col[key] == 1):
                vals.add(key)
        res.append(vals)
    ress = set()
    for x in res:
        if len(ress) == 0:
            ress = x
        else:
            ress.intersection(x)
    return ress
        
def derivG(ctx, G1):
    res = set()
    rows = []
    for g in G1:
        rows.append(ctx.getRowByG(g))
    for m in ctx.M:
        temp = True
        for row in rows:
            if (row[m] == 0):
                temp = False
                break
        if temp:
            res.add(m)
    return res
def derivM(ctx, M1):
    res = set()
    cols = []
    for m in M1:
        cols.append(ctx.getColByM(m))
    for g in ctx.G:
        temp = True
        for col in cols:
            if col[g] == 0:
                temp = False
                break;
        if temp:
            res.add(g)
    return res

def nextNeighbours(ctx):
    C = [Node(ctx.G, derivG(ctx, ctx.G))]
    E = set()
    currentLevel = set({Node(ctx.G, derivG(ctx, ctx.G))})
    while len(currentLevel) != 0:
        print(currentLevel)
        nextLevel = set()
        for node in currentLevel:
            lowerNeighbours = findLowerNeighbours(node, ctx)
            for lowerNode in lowerNeighbours:
                if lowerNode not in C:
                    C.append(lowerNode)
                    nextLevel.add(lowerNode)
                E.add((lowerNode.__hash__(), node.__hash__())) # add edge loweNode -> node to E
        currentLevel = nextLevel
    return Lattice(C, E)
                
            

def maxGen(candidates):
    res = set()
    for c in candidates:
        general = True
        for cc in candidates:
            if c == cc:
                continue
            if len(cc.G.intersection(c.G)) == len(c.G):
                general = False
                break
        if general:
            res.add(c)
    return res

def findLowerNeighbours(node, ctx):
    candidates = set()
    for m in ctx.M:
        if m in node.M:
            continue
        x1 = node.M.copy()
        x1.add(m)
        x1 = derivM(ctx, x1)
        y1 = derivG(ctx, x1)
        if (Node(x1, y1) not in candidates):
            candidates.add(Node(x1, y1))
    return maxGen(candidates)
        
```
